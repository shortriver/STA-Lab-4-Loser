I feel the need to explain that it is not my intention to represent Brazil as some murderous wasteland for LGBT+ people, that would be entirely too simplistic and disrespectful. The fact of the matter is that Brazil is a diverse and complex place, and in that place there are some people who murder LGBT+ folks, their allies, and cultural rhetoric is sometimes so hateful/unaccepting that LGBT+ people are compelled to take their own lives. As an outsider, I'm also compelled to mention that *obviously* this is the case in the United States as well, and none of this is mentioned in an attempt to hop on a cultural high-horse. This is the data I have, these are friends of my friends, family of my friends, and siblings within the LGBT+ community. 

So, what is the data? The people at [Grupo Gay da Bahia](http://www.ggb.org.br/ggb-ingles.html) provide the only sourced [record](https://homofobiamata.wordpress.com/)  of LGBT+ murders for the entire country of Brazil.

The data has problems: they can only report what is verifiable by news, legal documents, or the social media profiles of victims. People caught in between/outside binaries (bisexual, pansexual, nonbinary genders) are often hard to detect, or a family's shame will prevent accurate reporting of a victim's LGBT+ identity. Similarly, transgender, transexual, and transvestite Brazilians may also be misgendered by media, or their families' intervention.

That said, GGB's list still compiles enough data that we know that Brazil has the highest number of murders of Trans* *individuals in the world. Due to a very high level of violence throughout the country, sometimes it can be hard to distinguish whether or not that's just for being LGBT+ or for being Brazilian. To minimize some of the confusion, I focused the data to three metropolitan areas in Brazil: São Paulo, Rio de Janeiro, and the Distrito Federal. Coincidentally, these are the 1st, 2nd, and 4th densest areas in the country (see table). While murder rates are higher in denser areas, they are not disproportionately so compared to many of the Northeastern regions of Brazil. In the case of Brasília, or the DF, I'd actually argue they're ~~suspiciously~~ surprisingly low, and have a few ideas as to why that may be.
